```python
df = pd.read_csv("data.csv")
df.head(3)
```

```python
x=df.iloc[:4,0].values
y=df.iloc[:4,3].values
z=df.iloc[:4,1].values
q=df.iloc[:4,2].values
```

```python
y
```

```python
x
```

```python
plt.plot(x,y,linestyle='--')  # linestyle='-' par defaut  (pour un line ------)
plt.title("TV vs Sales")
plt.xlabel("TV or Radio")
plt.ylabel("Sales")
plt.plot(z,y,marker='o')  # pour des cyrcle plain
# methode 1 
plt.legend(['TV vs vente ','Radio vs vente']) # utilise cette fonction pour marque chaque droite (l'ordre obligatoire)
# methode 2
#plt.plot(x,y,label='TV vs vente') 
#plt.plot(z,y,color='r',label='Radio vs vente')
plt.grid() # pour les carres
plt.savefig("ahmed.png") # pour enregestrement
plt.show() # pour meme graphe 
# Si nous écrivons quelque chose. l'afficher sur un autre graphique .
```

# Bar Charts 

```python
#plusieurs fonction comme plot mais des args différent
```

```python
width = 0.25 #des idées
plt.bar(x - width,y,color="g",width=width,label='TV vs vente')# obligatoir plt.legend() (pour label )
plt.bar(x + width,z,color="b",width=width, label='Radio vs vente')
plt.bar(x,q,color="r",width=width, label='Newspaper vs vente')
plt.title("TV vs Sales")
plt.xlabel("TV or Radio")
plt.ylabel("Sales")
plt.xticks(ticks=x,label=y) # pour les valeur de l'axe des abscisses
plt.legend() 
plt.show() 
```
