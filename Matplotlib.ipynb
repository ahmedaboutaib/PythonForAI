# Bar Charts 

```python
#plusieurs fonction comme plot mais des args différent
#plt.xticks(ticks=x,label=y) # pour les valeur de l'axe des abscisses
```

```python
width = 0.25 #des idées
plt.bar(x - width,y,color="g",width=width,label='TV vs vente')# obligatoir plt.legend() (pour label )
plt.bar(x + width,z,color="b",width=width, label='Radio vs vente')
plt.bar(x,q,color="r",width=width, label='Newspaper vs vente')
plt.title("TV vs Sales")
plt.xlabel("TV or Radio")
plt.ylabel("Sales")
plt.xticks(ticks=x,label=y) # pour les valeur de l'axe des abscisses
plt.legend() 
plt.show() 
```

# Histograms

```python
import numpy as np
#plt.hist(ages,bins=bins ,edgecolor='r') 
#plt.axvline(moy,color='g',linewidth=3)  On n'utilise pas cette fonction pour tracer une ligne verticale pour la moyenne.
#admin.index # Pour récupérer l'index d'un DataFrame dans la bibliothèque pandas.
```

```python
ages= [12,13,14,15,23,43,21,56,78,97,22,11,33,12,32,43,19,16,18,55,66,77,88,99,40,50,60]
#plt.hist(ages,bins=len(ages) ,edgecolor='r') # edgecolor : pour les bords , bins : nembre des bacs 
bins=[i for i in range(10,110,10)] # Diviser l'histogramme en plusieurs parties.
plt.hist(ages,bins=bins ,edgecolor='r') 
moy=np.mean(ages)
plt.axvline(moy,color='g',linewidth=3) # On n'utilise pas cette fonction pour tracer une ligne verticale pour la moyenne.
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequence')
plt.show()
```

```python
dd=pd.read_csv('advertising.csv')
dd.head()
```

```python
x=dd['TV']
x.head()
```

```python
y=dd['Sales']
y.head()
```

```python
#plt.scatter(x,y,color='g',s=10,edgecolor='r',linewidth=3,alpha=0.5) # edgecolor:  bord ,alpha: transparnce ,
colors=x
size=x
plt.scatter(x,y, c=colors,s=size,edgecolor='r',linewidth=3,alpha=0.5,cmap='Reds') #  c  : l'intendite de la couloeur change en fonction de la valeur
# s :size => comment color
#cmap : chose le couleur de la faison (tadaroje) (rechercher en google sur : colormaps)
plt.colorbar().set_label('color bar') #  Une barre pour les différentes couleurs (scatter)
plt.show() 
```
